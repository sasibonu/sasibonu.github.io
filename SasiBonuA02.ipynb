```python
import numpy as np
import pandas as pd
import re
#import nltk
#from nltk.stem.porter import PorterStemmer
from datasets import load_dataset
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

#from nltk.corpus import stopwords
```

```python
dataset_en_hi = load_dataset("open_subtitles", "en-hi")
dataset_da_ru = load_dataset("open_subtitles", "da-ru")
```

```python
data_en_hi = dataset_en_hi["train"]
data_da_ru = dataset_da_ru["train"]
```

```python
data_en_hi[0:2]
```

```python
from gensim.models import Word2Vec

english_texts = [item['en'] for item in data_en_hi['translation']]
hindi_texts = [item['hi'] for item in data_en_hi['translation']]


all_texts = english_texts + hindi_texts

token_sentences = [sentence.split() for sentence in all_texts]

word2vec_model = Word2Vec(sentences=token_sentences, vector_size=100, window=5, min_count=1, sg=1)

words = []
for sentence in token_sentences:
    vectors = [word2vec_model.wv[word] for word in sentence if word in word2vec_model.wv]
    if vectors:
        agg_vector = np.mean(vectors, axis=0)
        words.append(agg_vector)

word_vectors = np.array(words)

pca = PCA(n_components=2)
p_components = pca.fit_transform(word_vectors)

kmeans = KMeans(n_clusters=2)
kmeans.fit(p_components)
cluster_labels = kmeans.labels_


print("PCA Components:")
print(p_components)
print("Cluster Labels:")
print(cluster_labels)

plt.figure(figsize=(8, 6))
plt.scatter(p_components[:, 0], p_components[:, 1], c=cluster_labels, cmap='viridis', edgecolor='k', s=100)
plt.title('K-means Clustering of Translations')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.grid(True)
plt.show()
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(p_components)
cluster_labels = kmeans.labels_

print("Cluster Labels:")
print(cluster_labels)

plt.figure(figsize=(8, 6))
plt.scatter(p_components[:, 0], p_components[:, 1], c=cluster_labels, cmap='viridis', edgecolor='k', s=100)
plt.title('K-means Clustering of Translations')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.grid(True)
plt.show()
```

```python
import numpy as np
from gensim.models import Word2Vec
from sklearn.decomposition import PCA
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

english_texts = [item['en'] for item in data_en_hi_red['translation']]
hindi_texts = [item['hi'] for item in data_en_hi_red['translation']]

all_texts = english_texts + hindi_texts

token_sentences = [sentence.split() for sentence in all_texts]

word2vec_model = Word2Vec(sentences=token_sentences, vector_size=100, window=5, min_count=1, sg=1)

words = []
for sentence in token_sentences:
    vectors = [word2vec_model.wv[word] for word in sentence if word in word2vec_model.wv]
    if vectors:
        agg_vector = np.mean(vectors, axis=0)
        words.append(agg_vector)

word_vectors = np.array(words)

pca = PCA(n_components=2)
principal_components = pca.fit_transform(word_vectors)

hierar_clust = linkage(principal_components, 'ward')

plt.figure(figsize=(10, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Samples')
plt.ylabel('Distance')
dendrogram(hierar_clust, labels=all_sentences, leaf_rotation=90)
plt.show()
```
